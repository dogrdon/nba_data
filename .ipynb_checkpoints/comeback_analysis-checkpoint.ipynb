Now we want to create columns for whether the home team one or not and for each event whether they were ahead or not. This is easy because if the SCOREMARGIN is Negative, the home team is losing. And if the last entry for a GAME_ID score margin is negative, then the home team lost.

```python
final_scores = dropped.groupby('GAME_ID').tail(n=1)
```

```python
final_scores.head()
```

```python
def home_adv(row):
    if row['SCOREMARGIN'] == 'TIE':
        return -1
    if int(row['SCOREMARGIN']) < 0:
        return 0
    if int(row['SCOREMARGIN']) > 0:
        return 1
    return -999
```

```python
final_scores['home_win'] = final_scores.apply(lambda x: home_win(x), axis=1)
```

```python
dropped['home_ahead'] = dropped.apply(lambda x: home_adv(x), axis=1)
```

```python
wins_df = final_scores[['GAME_ID', 'home_win']]
```

```python
prepped_data = dropped.merge(wins_df, left_on='GAME_ID', right_on='GAME_ID', how='inner')
```
